```python
annee= {}
for ligne in df["date"]:
    #print(ligne)
    d=ligne.split(', ')
    
    if (annee.get(d[1])==None):
        annee.update({d[1]:1})
    else :
        annee[d[1]]+=1
    
#print(annee)
```

```python

#Creation axe des abssice
xpt = [int(k)  for  k in  annee.keys()]
xpt.reverse()

#creation axe ordonnes
ypt = [k  for  k in  annee.values()]
ypt.reverse()
print(xpt)
print(ypt)

plt.plot(xpt, ypt)
plt.show()
```
